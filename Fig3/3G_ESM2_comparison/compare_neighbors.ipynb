### compute neighbor overlap

```python
def _shared_genes(m1: pd.DataFrame, m2: pd.DataFrame) -> List[str]:
    """
    Genes must exist in BOTH index and columns for each matrix (square similarity matrix).
    Returns sorted shared genes.
    """
    g1 = set(m1.index) & set(m1.columns)
    g2 = set(m2.index) & set(m2.columns)
    shared = sorted(g1 & g2)
    if len(shared) == 0:
        raise ValueError("No shared genes found between the two matrices (considering both index and columns).")
    return shared


def _restrict_and_align(m: pd.DataFrame, genes: List[str]) -> pd.DataFrame:
    """Restrict to shared genes and enforce same ordering on rows/cols."""
    return m.loc[genes, genes]


def top_neighbors(
    sim: pd.DataFrame,
    top_n: int = 20,
    *,
    exclude_self: bool = True,
    dropna: bool = True,
) -> Dict[str, List[str]]:
    """
    For each gene (row), return the top_n most similar genes (by value).
    Assumes rows/cols are genes, but does not require symmetry.
    """
    if top_n <= 0:
        raise ValueError("top_n must be > 0")

    genes = sim.index.to_numpy()
    arr = sim.to_numpy(dtype=float, copy=True)

    # Handle NaNs
    if dropna:
        arr = np.where(np.isnan(arr), -np.inf, arr)

    # Optionally exclude self-similarity
    if exclude_self:
        np.fill_diagonal(arr, -np.inf)

    k = min(top_n, max(arr.shape[1] - (1 if exclude_self else 0), 0))
    if k == 0:
        return {g: [] for g in genes}

    # Get indices of top-k per row (fast)
    idx_part = np.argpartition(arr, -k, axis=1)[:, -k:]
    row_idx = np.arange(arr.shape[0])[:, None]
    vals = arr[row_idx, idx_part]
    order = np.argsort(vals, axis=1)[:, ::-1]  # descending
    top_idx = idx_part[row_idx, order]

    neighbors = {}
    for i, g in enumerate(genes):
        nbrs = genes[top_idx[i]].tolist()
        # If dropna=False, you might have NaNs left; filter any -inf that came from NaN exclusion
        if dropna:
            nbrs = [x for x in nbrs if np.isfinite(arr[i, np.where(genes == x)[0][0]])]
        neighbors[g] = nbrs[:k]
    return neighbors


@dataclass(frozen=True)
class OverlapResult:
    neighbor_lists_1: Dict[str, List[str]]
    neighbor_lists_2: Dict[str, List[str]]
    overlap_table: pd.DataFrame


def neighbor_overlap_between_matrices(
    m1: pd.DataFrame,
    m2: pd.DataFrame,
    *,
    top_n: int = 20,
    exclude_self: bool = True,
    dropna: bool = True,
) -> OverlapResult:
    """
    1) Find shared genes
    2) Compute top-N neighbors for each gene in each matrix
    3) Compute overlap statistics per gene

    Returns:
      - neighbor_lists_1, neighbor_lists_2: dict gene -> list of neighbors
      - overlap_table: DataFrame indexed by gene with overlap metrics
    """
    shared = _shared_genes(m1, m2)
    a = _restrict_and_align(m1, shared)
    b = _restrict_and_align(m2, shared)

    n1 = top_neighbors(a, top_n=top_n, exclude_self=exclude_self, dropna=dropna)
    n2 = top_neighbors(b, top_n=top_n, exclude_self=exclude_self, dropna=dropna)

    records = []
    for g in shared:
        s1, s2 = set(n1[g]), set(n2[g])
        inter = s1 & s2
        union = s1 | s2

        n1_count = len(s1)
        n2_count = len(s2)
        inter_count = len(inter)
        union_count = len(union)

        jaccard = inter_count / union_count if union_count else np.nan
        overlap_coeff = inter_count / min(n1_count, n2_count) if min(n1_count, n2_count) else np.nan

        records.append(
            {
                "gene": g,
                "n_neighbors_m1": n1_count,
                "n_neighbors_m2": n2_count,
                "intersection_size": inter_count,
                "union_size": union_count,
                "jaccard": jaccard,
                "overlap_coefficient": overlap_coeff,
                "shared_neighbors": sorted(inter),
            }
        )

    overlap_df = pd.DataFrame.from_records(records).set_index("gene")
    return OverlapResult(neighbor_lists_1=n1, neighbor_lists_2=n2, overlap_table=overlap_df)


# -----------------------
# Example usage:
# -----------------------
# m1 and m2 are your pairwise similarity matrices as pandas DataFrames
# result = neighbor_overlap_between_matrices(m1, m2, top_n=50)
# print(result.overlap_table.sort_values("jaccard", ascending=False).head(10))
# neighbors_for_geneX_in_m1 = result.neighbor_lists_1["GENE_X"]
# neighbors_for_geneX_in_m2 = result.neighbor_lists_2["GENE_X"]
```

```python
from __future__ import annotations

import numpy as np
import pandas as pd


def _fd_bin_edges(x: np.ndarray) -> np.ndarray:
    """
    Freedman–Diaconis binning.
    Returns bin edges spanning [min, max]. Falls back to 10 bins if degenerate.
    """
    x = x[np.isfinite(x)]
    if x.size == 0:
        return np.array([0.0, 1.0])

    x_min, x_max = float(np.min(x)), float(np.max(x))
    if x_min == x_max:
        # All values identical
        return np.array([x_min, x_max + 1e-12])

    q25, q75 = np.percentile(x, [25, 75])
    iqr = q75 - q25
    if iqr <= 0:
        # fallback
        return np.linspace(x_min, x_max, 11)

    bin_width = 2 * iqr * (x.size ** (-1 / 3))
    if bin_width <= 0:
        return np.linspace(x_min, x_max, 11)

    n_bins = int(np.ceil((x_max - x_min) / bin_width))
    n_bins = max(n_bins, 1)
    return np.linspace(x_min, x_max, n_bins + 1)


def summarize_jaccard_bins(
    overlap: pd.DataFrame | pd.Series,
    *,
    jaccard_col: str = "jaccard",
    method: str = "linspace_0_1",  # "manual", "linspace_0_1", "linspace_data", "fd", "quantile"
    n_bins: int = 10,
    bin_edges: list[float] | np.ndarray | None = None,  # used for method="manual"
    dropna: bool = True,
    right: bool = True,
    include_lowest: bool = True,
) -> pd.DataFrame:
    """
    Summarize how many genes fall into Jaccard bins.

    Parameters
    ----------
    overlap : DataFrame or Series
        If DataFrame: uses overlap[jaccard_col]. If Series: uses it directly.
    method : str
        - "manual": use bin_edges you pass in
        - "linspace_0_1": uniform bins from 0 to 1 (good for Jaccard)
        - "linspace_data": uniform bins from min(x) to max(x)
        - "fd": Freedman–Diaconis bins (data-driven width)
        - "quantile": equal-frequency bins (pd.qcut)
    n_bins : int
        Number of bins (used for linspace_* / quantile; fd chooses its own)
    bin_edges : list/ndarray
        Required for method="manual"
    """
    if isinstance(overlap, pd.DataFrame):
        x = overlap[jaccard_col].astype(float)
    else:
        x = overlap.astype(float)

    if dropna:
        x = x.dropna()

    if method == "quantile":
        # qcut can fail if many identical values; duplicates="drop" prevents crashing
        cats = pd.qcut(x, q=min(n_bins, max(len(x), 1)), duplicates="drop")
        binned = cats

    else:
        if method == "manual":
            if bin_edges is None:
                raise ValueError('method="manual" requires bin_edges=[...]')
            edges = np.asarray(bin_edges, dtype=float)

        elif method == "linspace_0_1":
            edges = np.linspace(0.0, 1.0, n_bins + 1)

        elif method == "linspace_data":
            if len(x) == 0:
                edges = np.array([0.0, 1.0])
            else:
                edges = np.linspace(float(x.min()), float(x.max()), n_bins + 1)

        elif method == "fd":
            edges = _fd_bin_edges(x.to_numpy())

        else:
            raise ValueError(f"Unknown method: {method}")

        binned = pd.cut(
            x,
            bins=edges,
            right=right,
            include_lowest=include_lowest,
        )

    out = (
        binned.value_counts(sort=False, dropna=False)
        .rename("count")
        .to_frame()
        .assign(
            fraction=lambda d: d["count"] / d["count"].sum() if d["count"].sum() else np.nan,
            cumulative_count=lambda d: d["count"].cumsum(),
            cumulative_fraction=lambda d: d["fraction"].cumsum(),
        )
        .reset_index()
        .rename(columns={"index": "bin"})
    )
    return out

```
